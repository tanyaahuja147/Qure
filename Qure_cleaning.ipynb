```python
# import seaborn as sns
# sns.countplot(x="county", data=mobility)
```

```python
#Mode of each county per state
temp = mobility.groupby(['state'])['county'].agg(lambda x: pd.Series.mode(x)[0]).to_frame()
temp.rename(columns={"state": "state_mode", "county": "county_mode"},inplace=True)
temp
```

```python
mobility=mobility.merge(temp, how = 'inner', on = 'state')
```

```python
#Replaced value 'Total' with each states most common county
import numpy as np
mobility['county'] = np.where(mobility['county'] != 'Total', mobility['county'], mobility['county_mode'])
```
